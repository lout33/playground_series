```python
def f_cost_2(x,y,m,b):
  # implement the cost function from scratch
  # return 14*m**2 + 3*b**2 + 38 + 12*m*b - 42*m - 20*b
  c = x.shape[0]
  total_cost = 0
  for i in range(c):
    f_mb = m*x[i] + b
    cost = (f_mb - y[i])**2
    # cost = abs(f_mb - y[i])

    total_cost += cost

  return total_cost
```

```python
def f_cost_2(x,y,m,b):
  c = x.shape[0]
  total_cost = 0
  for i in range(c):
    f_mb = m*x[i] + b
    cost = (f_mb - y[i])**2
    # cost = abs(f_mb - y[i])
    total_cost += cost

  # return total_cost
  return (1/(2*c)) * total_cost
```

```python
print(f_cost_2(x,y,2,2))
```

```python
def derivada_wrt_m_b(x,y,m,b):
  c = x.shape[0]
  dj_dm = 0
  dj_db = 0
  for i in range(c):
    # derivative with respect to m
    # dj_dm_i = 2 * (x[i]*m+b - y[i]) * x[i]
    dj_dm_i = ((x[i]*m+b - y[i]) * x[i]) / c
    # derivative with respect to b
    # dj_db_i = 2 * (x[i]*m+b - y[i])
    dj_db_i =  ((x[i]*m+b - y[i])) / c

    dj_dm = dj_dm + dj_dm_i
    dj_db = dj_db + dj_db_i

  return dj_dm, dj_db
```

```python
res1, res2 = derivada_wrt_m_b(X,Y,2,2)
print(res1)
print(res2)
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, derivada_wrt_m_b, f_cost_2):
  # initialize the weights
  w_out = w_in
  b_out = b_in
  J_history = []
  p_history = []

  for i in range(num_iters):
  # calculate the gradient
    dj_dm, dj_db = derivada_wrt_m_b(x, y, w_out, b_out)
  # update the weights
    w_out = w_out - alpha * dj_dm
    b_out = b_out - alpha * dj_db
  
    J_history.append(f_cost_2(x, y, w_out , b_out))
    p_history.append([w_out, b_out])
    if i% math.ceil(num_iters/10) == 0:
      print(f"Iteration {i:4}: Cost {float(J_history[-1]):8.2f}   ")
  return w_out, b_out, J_history, p_history
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import copy, math
```

```python
X_train = np.array([[1, 2],
                    [2, 3],
                    [6, 7],
                    [11, 12]])

# X_train = np.array([[1, 2, 6, 11],
#                     [2, 3, 7, 12]])
y_train = np.array([2, 4, 8, 10])
b_init = 0
w_init = np.array([0.5, 0.5])
```

```python
def cost_function(x, y, m, b):
    c = x.shape[0]
    total_cost = 0
    for i in range(c):
        # print(x[i])
        # print(m)
        f_mb = np.dot(x[i] , m) + b
        cost = (f_mb - y[i])**2
        # cost = abs(f_mb - y[i])
        total_cost = total_cost + cost

    total_cost = total_cost / (2 * c)    #scalar
    return total_cost

print(cost_function(X_train, y_train, w_init, b_init))
```

```python
def multiple_derivada_wrt_m_b(x,y,m,b):
  o,p = x.shape           #(number of examples, number of features)
  dj_dm = np.zeros((p,))
  dj_db = 0

  for i in range(o):
    dj_db_i = (np.dot(x[i], m) + b) - y[i]
    for j in range(p):
      dj_dm_i = ((np.dot(x[i], m) + b) - y[i]) * x[i,j]
      dj_dm[j] = dj_dm[j] + dj_dm_i
    dj_db = dj_db + dj_db_i

  dj_dm = dj_dm / o
  dj_db = dj_db / o
  return dj_dm, dj_db

res1, res2 = multiple_derivada_wrt_m_b(X_train,y_train,w_init,b_init)
print(res1)
print(res2)
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, derivada_wrt_m_b,cost_function):
  # initialize the weights
  w_out = copy.deepcopy(w_in)
  b_out = b_in
  J_history = []
  # p_history = []

  for i in range(num_iters):
  # calculate the gradient
    dj_dm, dj_db = derivada_wrt_m_b(x, y, w_out, b_out)
  # update the weights
    w_out = w_out - alpha * dj_dm
    b_out = b_out - alpha * dj_db
    J_history.append(cost_function(x, y, w_out , b_out))

    # p_history.append([w_out, b_out])
    if i% math.ceil(num_iters/10) == 0:
      print(f"Iteration {i:4}: Cost {float(J_history[i])}   ")

  # return w_out, b_out, J_history, p_history
  # return w_out, b_out, J_history
  return w_out, b_out
```

```python
w, b = gradient_descent(X_train,y_train,w_init,b_init,0.004,3000,multiple_derivada_wrt_m_b,cost_function)
# round to 2 decimal places
w = np.round(w, 2)
b = np.round(b, 2)
print(w, b)

# [-0.19  0.97] 1.16
```
