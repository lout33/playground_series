```python
def f_cost_2(x,y,m,b):
  c = x.shape[0]
  total_cost = 0
  for i in range(c):
    f_mb = m*x[i] + b
    cost = (f_mb - y[i])**2
    # cost = abs(f_mb - y[i])
    total_cost += cost

  # return total_cost
  return (1/(2*c)) * total_cost
```

```python
def derivada_wrt_m_b(x,y,m,b):
  c = x.shape[0]
  dj_dm = 0
  dj_db = 0
  for i in range(c):
    # derivative with respect to m
    # dj_dm_i = 2 * (x[i]*m+b - y[i]) * x[i]
    dj_dm_i = ((x[i]*m+b - y[i]) * x[i]) / c
    # derivative with respect to b
    # dj_db_i = 2 * (x[i]*m+b - y[i])
    dj_db_i =  ((x[i]*m+b - y[i])) / c

    dj_dm = dj_dm + dj_dm_i
    dj_db = dj_db + dj_db_i

  return dj_dm, dj_db
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, derivada_wrt_m_b, f_cost_2):
  # initialize the weights
  w_out = w_in
  b_out = b_in
  J_history = []
  p_history = []

  for i in range(num_iters):
  # calculate the gradient
    dj_dm, dj_db = derivada_wrt_m_b(x, y, w_out, b_out)
  # update the weights
    w_out = w_out - alpha * dj_dm
    b_out = b_out - alpha * dj_db
  
    J_history.append(f_cost_2(x, y, w_out , b_out))
    p_history.append([w_out, b_out])
    if i% math.ceil(num_iters/10) == 0:
      print(f"Iteration {i:4}: Cost {float(J_history[-1]):8.2f}   ")
  return w_out, b_out, J_history, p_history
```

```python
def cost_function(x, y, m, b):
    c = x.shape[0]
    total_cost = 0
    for i in range(c):
        # print(x[i])
        # print(m)
        f_mb = np.dot(x[i] , m) + b
        cost = (f_mb - y[i])**2
        # cost = abs(f_mb - y[i])
        total_cost = total_cost + cost

    total_cost = total_cost / (2 * c)    #scalar
    return total_cost

print(cost_function(X_train, y_train, w_init, b_init))
```

```python
def multiple_derivada_wrt_m_b(x,y,m,b):
  o,p = x.shape
  # print(x.shape)       #(number of examples, number of features)
  dj_dm = np.zeros((p,))
  dj_db = 0

  for i in range(o):
    dj_db_i = (np.dot(x[i], m) + b) - y[i]
    for j in range(p):
      dj_dm_i = ((np.dot(x[i], m) + b) - y[i]) * x[i,j]
      dj_dm[j] = dj_dm[j] + dj_dm_i
    dj_db = dj_db + dj_db_i

  dj_dm = dj_dm / o
  dj_db = dj_db / o
  return dj_dm, dj_db

res1, res2 = multiple_derivada_wrt_m_b(X_train,y_train,w_init,b_init)
print(res1)
print(res2)
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, derivada_wrt_m_b,cost_function):
  # initialize the weights
  w_out = copy.deepcopy(w_in)
  b_out = b_in
  J_history = []
  # p_history = []

  for i in range(num_iters):
  # calculate the gradient
    dj_dm, dj_db = derivada_wrt_m_b(x, y, w_out, b_out)
  # update the weights
    w_out = w_out - alpha * dj_dm
    b_out = b_out - alpha * dj_db
    J_history.append(cost_function(x, y, w_out , b_out))

    # p_history.append([w_out, b_out])
    if i% math.ceil(num_iters/10) == 0:
      print(f"Iteration {i:4}: Cost {float(J_history[i])}   ")

  # return w_out, b_out, J_history, p_history
  # return w_out, b_out, J_history
  return w_out, b_out
```

```python
# create target data
x = np.arange(0, 20, 1)
print(x)

y = 0.1 * (x) + 5
# convert to numpy array y
y = np.array(y)
print(y)

plt.scatter(x, y, marker='x', c='r', label="Actual Value"); plt.title("no feature engineering")
```

```python
X_train = np.array([x]).T
y_train = np.array([y]).T
print(X_train.shape)
print(y_train.shape)
```

```python
scaler = StandardScaler()
X_norm = scaler.fit_transform(X_train)
print(f"Peak to Peak range by column in Raw        X:{np.ptp(X_train,axis=0)}")   
print(f"Peak to Peak range by column in Normalized X:{np.ptp(X_norm,axis=0)}")
```

```python
sgdr = SGDRegressor(max_iter=1000)
sgdr.fit(X_norm, y_train)
print(sgdr)
print(f"number of iterations completed: {sgdr.n_iter_}, number of weight updates: {sgdr.t_}")
```

```python
b_norm = sgdr.intercept_
w_norm = sgdr.coef_
print(f"model parameters:                   w: {w_norm}, b:{b_norm}")
print( "model parameters from previous lab: w: [110.56 -21.27 -32.71 -37.97], b: 363.16")
```

```python
# make a prediction using sgdr.predict()
y_pred_sgd = sgdr.predict(X_norm)
# make a prediction using w,b.
y_pred = np.dot(X_norm, w_norm) + b_norm
print(f"prediction using np.dot() and sgdr.predict match: {(y_pred == y_pred_sgd).all()}")

print(f"Prediction on training set:\n{y_pred[:4]}" )
print(f"Target values \n{y_train[:4]}")
```

```python
# plot predictions and targets vs original features    
X_features = ['size(sqft)']
fig,ax=plt.subplots(1,2,figsize=(12,3),sharey=True)
for i in range(len(ax)):
    ax[i].scatter(X_train[:,i],y_train, label = 'target')
    ax[i].set_xlabel(X_features[i])
    ax[i].scatter(X_train[:,i],y_pred, label = 'predict')
ax[0].set_ylabel("Price"); ax[0].legend();
fig.suptitle("target versus prediction using z-score normalized model")
plt.show()
```
